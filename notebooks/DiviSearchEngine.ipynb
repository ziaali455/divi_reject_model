This file does the following: it loads in data from a json with company name,CNAE,address,foudnation date, and website. It strips the date added down to a simple integer, and checks if the CNAE is within a range and replaces the value with a string A-Z instead. This information is saved to a JSON which is later meant to be used in a C# and .NET file that takes in the information 

```python
import pandas as pd
import numpy as np
import json
json_file_path = "/Users/alizia/AliDivi/accounts.json"
df = pd.read_json(json_file_path)
# Load the JSON data
with open('/Users/alizia/AliDivi/accounts.json') as json_file:
    data = json.load(json_file)
# Normalize the nested JSON column
arr = data['data']

df = pd.DataFrame.from_dict(arr['results'])
df

# styled_df = df.style \
#     .set_properties(**{'text-align': 'center'}) \
#     .set_table_styles([{
#         'selector': 'th',
#         'props': [('background-color', '#000000'), ('font-weight', 'bold')]
#     }])

# styled_df
```

```python
newDF = df[['nickName','activityCodes', 'foundationDate']].copy()
newDF
```

```python
newDF = newDF.drop_duplicates(subset=['nickName'])
newDF
```

```python
filteredDF = newDF[newDF.astype(str)['activityCodes'] != '[]']
filteredDF['foundationDate'] = filteredDF['foundationDate'].str.slice(0,10)

filteredDF
```

```python
listOfLists = []
for lists in filteredDF['activityCodes']:
    myList = []
    for string in lists:
        string = string[0:2]
        if string=='':
            myList.append(-1)
        else:
            myList.append(int(string))
    listOfLists.append(myList)
listOfLists
```

```python
filteredDF['activityCodes'] = listOfLists
```
