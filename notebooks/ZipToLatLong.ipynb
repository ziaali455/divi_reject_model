Creating a function to only add to a list of cityState pairs if the address field was filled allowed me to circumvent this issue:

```python
#I'm not fully clear on how this works
def read_json() -> list[dict]:
    with open ("accounts.json", encoding="utf-8") as f:
        data = json.loads(f.read())
        data = data ["data"] ["results"]
        return [f for f in data if len(f["addresses"]) > 0]
```

```python
raw = read_json()
df = pd.DataFrame.from_dict(raw)
df
```

```python
cityStatePairs = []
for company in raw:
    for address in company['addresses']:
        if address['city'] and address['stateAbbreviation']:
            cityStatePairs.append(address['city'] + " " + address['stateAbbreviation'])
        else:
            cityStatePairs.append(None)

df['cityState'] = cityStatePairs
df = df[['companyName','cityState']]
df
```

Now that we've achieved the proper city and state pairs, we can start making attempts to convert this pair into an actual latitude and longitude, and then store this info in another column on the table. This process can be done for any set of company data formatted in the same way as on the Contas page.

```python
latLong = []
import geopy
import geopandas
from geopy import Nominatim
locator = Nominatim(user_agent='myGeocoder') # this is the line causing the issue in the else
location = locator.geocode('MARÍLIA SP')
print(str(location.latitude) + ", " + str(location.longitude))

for item in df['cityState']:
    if(item==None):
        latLong.append(None)
    else:
        location = locator.geocode(item)
        latLong.append(str(location.latitude) + ", " + str(location.longitude))
```

```python
import pandas as pd
df = pd.read_json('/Users/alizia/AliDivi/proof-of-concept/datasets/companyCoords.json')
crs = {'init':'epsg:4326'}
df["Latitude"] = df['coordinates'].str.split(',').str[0]
df["Longitude"] = df['coordinates'].str.split(',').str[1]
df = df[df.notnull()]
df
```

```python
geometry = [Point(xy) for xy in zip(df['Longitude'],df['Latitude'])]
geo_df = gpd.GeoDataFrame(df, #specify our data
                          crs=crs, #specify our coordinate reference system
                          geometry=geometry) #specify the geometry list we created
geo_df.head()
```

```python
fig, ax = plt.subplots(figsize=(15,15))
map.plot(ax=ax, alpha=0.4, color='grey')
geo_df.plot(ax=ax, 
                                       markersize=20, 
                                       color='blue', 
                                       marker='o', 
                                       label='Clients')

plt.legend(prop={'size':15})
```
